```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# optional: make plots prettier
sns.set(style="whitegrid")
%matplotlib inline
```

# üß™ Exploratory Data Analysis on McDonald's Menu Dataset

In this notebook, we will analyze the nutritional content of items on McDonald's menu.  
We aim to:
- Clean and understand the dataset
- Analyze stats like calories, fat, and sugar
- Group and visualize insights by item category
- Recommend actionable steps for health-focused menu changes


```python
df = pd.read_csv("menu.csv")
df.head()  # check the first few rows
df.info()  # check for missing values, data types
df.describe()  # summary statistics
df.isnull().sum()  # count missing values
```

```python
# Summary statistics
df.describe()

# Mode (most common value)
df.mode().iloc[0]
```

### üîç Observations from Descriptive Stats:
- Calories range widely across items
- Sugars and fats also vary significantly
- We'll now analyze how these differ by category


```python
# Mean calories, fat, and sugar per category
df.groupby("Category")[["Calories", "Total Fat", "Sugars"]].mean().round(2)
```

```python
# Histogram of Calories
plt.figure(figsize=(8, 5))
sns.histplot(df['Calories'], kde=True, bins=20)
plt.title("Distribution of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot: Calories by Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Calories')
plt.title("Calories by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot: Count of items per Category
plt.figure(figsize=(8, 4))
df['Category'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Number of Items per Category")
plt.ylabel("Count")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

## ‚úÖ Insights & Recommendations

- **Breakfast and Burgers** have the highest average calorie content.
- **Beverages** contain significantly higher sugar levels.
- Several categories have items exceeding 1,000 calories.
- Recommend:
  - Promoting lower-calorie or smaller-portion alternatives.
  - Highlighting low-sugar beverages.
  - Introducing balanced meal combinations with fruits or veggies.

This analysis can support more health-conscious menu planning at McDonald's.
